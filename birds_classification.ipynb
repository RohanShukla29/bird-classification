```
import os
from PIL import Image
from numpy import asarray
from os import listdir
from matplotlib import image
import numpy as np
```

```
train_data=list()
train_label=list()
label_index=0
for x in listdir('train/'):


  
  
    for filename in listdir('train/'+x+'/'):
        # load image

        #img_data = Image.open('jpg/'+x+'/'+filename)
        img_data=image.imread('train/'+x+'/'+ filename)

        # store loaded image
        #npa=asarray(img_data)
        npa=img_data
        train_data.append(npa)
        train_label.append(label_index)
        print('> loaded %s %s %s' % (filename, npa.shape, x))
    label_index=label_index+1
    
```

```
val_data=list()
val_label=list()
index_label=0
for x in listdir('validate/'):


  
  
    for filename in listdir('validate/'+x+'/'):
        # load image

        #img_data = Image.open('jpg/'+x+'/'+filename)
        img_data=image.imread('validate/'+x+'/'+ filename)

        # store loaded image
        #npa=asarray(img_data)
        npa=img_data
        val_data.append(npa)
        val_label.append(index_label)
        print('> loaded %s %s %s' % (filename, npa.shape, x))
    index_label=index_label+1
    
```

```
for i in range(0,len(val_label),1):
  if (i>=0 and i<5):
    val_label[i]=7
  if (i>=5 and i<10):
    val_label[i]=8
  if (i>=10 and i<15):
    val_label[i]=2
  if (i>=15 and i<20):
    val_label[i]=0
  if (i>=20 and i<25):
    val_label[i]=1
  if (i>=25 and i<30):
    val_label[i]=3
  if (i>=30 and i<35):
    val_label[i]=5
  if (i>=35 and i<40):
    val_label[i]=4
  if (i>=40 and i<45):
    val_label[i]=6
  if (i>=45 and i<50):
    val_label[i]=9

  print(val_label[i])
  
```

```
from keras.utils import to_categorical
train_labels_one_hot = to_categorical(train_label)
val_labels_one_hot = to_categorical(val_label)


print('Original label 0 : ', val_label[0])

print('After conversion to categorical ( one-hot ) : ', val_labels_one_hot[0])
```

```

train_data_np=np.array(train_data)
val_data_np=np.array(val_data)
```

```
import tensorflow as tf
from tensorflow import keras
print(tf.__version__)
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Flatten, Dense, InputLayer, BatchNormalization, Dropout
from keras.layers import Dense, Conv2D, MaxPooling2D, Dropout, Flatten
```

```
val_data_np.shape
```

```
train_data_np.shape
```

```
train_data_np = train_data_np.astype('float32')
val_data_np = val_data_np.astype('float32')
train_data_np /= 255
val_data_np /= 255
```

```
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.optimizers import SGD, Adam, RMSprop

model=Sequential()
model.add(Conv2D(32, (3,3),padding='same',input_shape=(224,224,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(10))
model.add(Activation('softmax'))
model.summary()

model.compile(loss='categorical_crossentropy', optimizer='RMSprop', metrics=['accuracy'])
model.fit(train_data_np, train_labels_one_hot, batch_size=4, epochs=34, verbose=1, validation_data=(val_data_np, val_labels_one_hot))
```

```
test_data=list()
test_label=list()
index_label=0
for x in listdir('test/'):


  
  
    for filename in listdir('test/'+x+'/'):
        # load image

        #img_data = Image.open('jpg/'+x+'/'+filename)
        img_data=image.imread('test/'+x+'/'+ filename)

        # store loaded image
        #npa=asarray(img_data)
        npa=img_data
        test_data.append(npa)
        test_label.append(index_label)
        print('> loaded %s %s %s' % (filename, npa.shape, x))
    index_label=index_label+1
    
```

```
for i in range(0,len(test_label),1):
  if (i>=0 and i<5):
    test_label[i]=6
  if (i>=5 and i<10):
    test_label[i]=7
  if (i>=10 and i<15):
    test_label[i]=0
  if (i>=15 and i<20):
    test_label[i]=3
  if (i>=20 and i<25):
    test_label[i]=8
  if (i>=25 and i<30):
    test_label[i]=1
  if (i>=30 and i<35):
    test_label[i]=9
  if (i>=35 and i<40):
    test_label[i]=5
  if (i>=40 and i<45):
    test_label[i]=4
  if (i>=45 and i<50):
    test_label[i]=2

  print(test_label[i])
```

```
test_data_np=np.array(test_data)
test_data_np = test_data_np.astype('float32')
test_data_np /= 255
```

```
test_labels_one_hot = to_categorical(test_label)
```

```
score=model.evaluate(test_data_np, test_labels_one_hot)
print('accuracy', score[1])
print('loss', score[0])
```

```
from sklearn.metrics import classification_report
y_pred=model.predict(test_data_np)
y_pred=np.argmax(y_pred, axis=1) 
print(y_pred)
print(asarray(test_label))
print(classification_report(test_label,y_pred))
```
